## Exercise 1: 
Create an abstract class called Vehicle with 
- color, make, and year private attributes 

and abstract method called 
- get_info() which returns all information
- honk() which returns a string of honking sound

Create Car, Boat, and Bicycle class which is a child class of the Vehicle class.  Then, create an instance of each child class and call get_info() method to print out its information in a user friendly format. Add other methods as you like.

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):
    def __init__(self, color, make, year): #not necessary to instantiate these
        self.__color = color
        self.__make = make
        self.__year = year
    
    @abstractmethod #letting Python know that this should be treated as an abstract method
    def get_info(self):
        pass #don't need to type it all out again
    
    @abstractmethod
    def honk(self):
        pass 

class Car(Vehicle):
    category = "Car"
    def __init__(self, color, make, year):
        self.__color = color
        self.__make = make
        self.__year = year
    
    def get_info(self):
        return f"{self.category} Color: {self.__color}\nMake: {self.__make}\nYear: {self.__year}"

    def honk(self):
        return "honk"

class Boat(Vehicle):
    category = "Boat"
    def __init__(self, color, make, year):
        self.__color = color
        self.__make = make
        self.__year = year
        
    def get_info(self):
        return f"Color: {self.category} {self.__color}\nMake: {self.__make}\nYear: {self.__year}"
    
    def honk(self):
        return "Boom"

class Bicycle(Vehicle):
    category = "Bicycle"
    def __init__(self, color, make, year):
        self.__color = color
        self.__make = make
        self.__year = year
    
    def get_info(self):
        return f"Color: {self.category} {self.__color}\nMake: {self.__make}\nYear: {self.__year}"
    
    def honk(self):
        return "ring"
    
car1 = Car("red", "toyota", 1999)
boat = Boat("white", "Chris Craft", 2010)
bicycle = Bicycle("red", "Huffy", 2021)

my_vehicles = [car1, boat, bicycle]

for vehicle in my_vehicles:
    print(vehicle.get_info())
    print(vehicle.honk())
    print("*" * 12)
    
```

## Exercise 2: 
Create an abstract class called Person with name and address attributes, and print_info() abstract method. 
1. Build an Employee class inherited from Person class that 
    - stores hired date and salary attributes 
    - print_info() method that will be used to print out all information about an employee
    - a method to calculate employee's monthly pay where federal tax is 15%, MD tax is 4.5%, and a local tax is 3%.
    
2. Build a Customer class also inherited from Person class that
    - stores balance attribute
    - print_info() method that will be used to print out all information about a customer


Build any method you need but apply encapsulation at all possible. Then, write a program to show your classes work with multiple employees and customers. 

CHALLENGE!!!
Create an Address class and use it in Person class

```python
from abc import ABC, abstractmethod

class Person(ABC):
    
    @abstractmethod
    def __init__(self, name, address):
        pass
    
    @abstractmethod
    def print_info(self):
        pass
    
    #code in our two attributes using "getter" methods
    def get_address(self):
        return self.__address
    
    def get_name(self):
        return self.__name

class Employee(Person):
    
    def __init__(self, name, address, hired_date, salary):
        self.__name = name
        self.__address = address 
        self.__hired_date = hired_date
        self.__salary = salary
    
    def print_info(self):
        return f"Name: \t\t{self.__name}\nAddress: \t\t{self.__address}\nHired Date: \t\t{self.__hired_date}\nSalary: \t\t{self.__salary}"
    
    def get_name(self):
        return self.__name
    
    def compute_monthly(self):
        monthly = self.__salary/12
        fedtax = 0.15 * monthly
        md_tax = 0.045 * monthly
        local_tax = 0.03 * monthly
        netpay = monthly - fedtax - md_tax - local_tax
        return f"Monthly pay before deductions: ${monthly:,.2f}\nMonthly pay after deductions: {netpay}"
    
class Customer(Person):
    def __init__(self, name, address, balance):
        self.__name = name
        self.__address = address
        self.__balance = balance
        
    def print_info(self):
        return f"Name: \t\t{self.__name}\nAddress: \t\t{self.__address}\nBalance: \t\t{self.__balance}"

e1 = Employee("Tyler", "Germantown", "June", "100000")
print(e1.get_name())

employee1 = Employee("Bugs Bunny", "WB Studios", 2010, 65000)
employee2 = Employee("Daffy Duck", "WB Studios", 2000, 85000)

customer1 = Customer("Big Bird", "Sesame Street", 200)
customer2 = Customer("Homer Simpson", "Springfield", 350)
customer3 = Customer("Oscar", "Sesame Street", 25)

print(employee1.print_info())
print(employee1.compute_monthly())
print("*" * 25)
print(employee2.print_info())
print(employee2.compute_monthly())

my_customers = [customer1, customer2, customer3]

for customer in my_customers:
    print(customer.print_info())
    print("*" * 25)
    
```
